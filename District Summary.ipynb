```python
import os
import csv
```

```python
import pandas as pd
```

```python
school_data_path = os.path.join("Resources", "schools_complete.csv")
student_data_path = os.path.join("Resources", "students_complete.csv")
print(school_data_path)
print(student_data_path)
```

```python
student_data_df = pd.read_csv(student_data_path)
student_data_df
```

```python
school_data_df = pd.read_csv(school_data_path)
school_data_df
```

```python
#MERGE SCHOOL & STUDENT DATA
school_data_complete = pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
school_total = school_data_complete["school_name"].value_counts()
school_total
#TOTAL SCHOOLS
school_total = school_total.count()
```

```python
student_total = school_data_complete.loc[:, ["student_name"]]
student_total
student_total.count()
#TOTAL STUDENTS
student_total = len(school_data_complete)
```

```python
budget_total = school_data_df.loc[:, ["budget"]]
budget_total
budget_total.sum()
#TOTAL BUDGET
total_budget = school_data_df["budget"].sum()
```

```python
#AVERAGE MATH SCORE
average_math_score = school_data_complete["math_score"].mean()
average_math_score
#AVERAGE READING SCORE
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
(average_math_score + average_reading_score)/2
#OVERALL PASSING SCORE
passing_score = (average_math_score + average_reading_score)/2
```

```python
school_data_complete["school_name"].head()
school_name = school_data_complete["school_name"].head()
#NEED A BOOLEAN TO CALCULATE HOW MANY STUDENTS ARE PASSING MATH\
#& READING
#.loc = location; thinking of the data like an Excel spreadsheet\
#(row, column); 
#can use labels/labeled index (ex. words)
#.iloc = numeric values (ex. numbers 0,1,2,3)
#both .loc & .iloc are followed by [rows,column]
school_data_complete.loc[school_data_complete["math_score"]>=70,\
                         "math_score"].count()/ student_total*100
```

```python
#PERCENTAGE PASSING MATH
passing_math = school_data_complete.loc[school_data_complete["math_score"]>=70,"math_score"].count()/ student_total*100
school_data_complete.loc[school_data_complete["reading_score"]>=70,"reading_score"].count()/ student_total*100
#PERCENTAGE PASSING READING
passing_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70,"reading_score"]\
.count()/ student_total*100
(passing_math + passing_reading)/2
#OVERALL PASSING SCORE
overall_passing_score = (passing_math + passing_reading)/2
```

### DISTRICT SUMMARY

```python
district_summary_df = pd.DataFrame({"Total Schools": [school_total],
                              "Total Students": [student_total],        
                              "Total Budget": [total_budget],
                              "Average Math Score": [average_math_score],
                              "Average Reading Score": [average_reading_score],
                              "% Passing Math": [passing_math],
                              "% Passing Reading": [passing_reading],
                              "% Overall Passing": [overall_passing_score]})
district_summary_df["Total Students"]=district_summary_df["Total Students"].map("{:,}".format)
district_summary_df["Total Budget"]=district_summary_df["Total Budget"].map("${:,.2f}".format)
district_summary_df
```

```python
pd.merge(student_data_df, school_data_df, how="left", on=["school_name", "school_name"])
```

```python
school_summary = school_data_df[["school_name","type","budget"]].set_index("school_name")
school_summary
```

```python
#GROUP BY
#How many times the school name appears = how many students per school
school_gb = school_data_complete.groupby("school_name")
school_summary["Total Students"] = school_gb.count()["type"]
school_summary["Per Student Budget"] = school_summary["budget"]/student_total
school_summary["Average Math Score"] = school_gb.mean()["math_score"]
school_summary["Average Reading Score"] = school_gb.mean()["reading_score"]
```

```python
total_passing_math = school_data_complete.loc[school_data_complete["math_score"]>=70]
total_passing_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70]
overall_passing_school = school_data_complete.loc[(school_data_complete["reading_score"]>=70)
                                                 &(school_data_complete["math_score"]>=70)]
```

```python
#PASSING STUDENTS
#GROUP-BY
school_summary["% Passing Math"] = total_passing_math.groupby("school_name")\
.count()["type"]/school_summary["Total Students"]*100
school_summary["% Passing Reading"] = total_passing_reading.groupby("school_name")\
.count()["type"]/school_summary["Total Students"]*100
school_summary["% Overall Passing Rate"] = overall_passing_school.groupby("school_name")\
.count()["type"]/school_summary["Total Students"]*100
```

### MATH SCORES BY GRADE

```python
math_scores_by_grades = school_data_complete.groupby(["school_name","grade"]).mean()["math_score"]
#SERIES w/ two indexes
#convert to a dataframe = pd.DataFrame()
#reset_index = reseting the index to a column (ex. 0, 1, 2)
math_scores_by_grades = pd.DataFrame(math_scores_by_grades).reset_index()
math_scores_by_grades = math_scores_by_grades.pivot(index = 'school_name', columns= 'grade',\
                                                          values = 'math_score')
#REORDER COLUMNS to put in order
math_scores_by_grades = math_scores_by_grades[["9th","10th","11th", "12th"]]
math_scores_by_grades
```

### READING SCORES BY GRADE

```python
reading_scores_by_grades = school_data_complete.groupby(["school_name","grade"]).mean()["reading_score"]
#SERIES w/ two indexes
#convert to a dataframe = pd.DataFrame()
#reset_index = reseting the index to a column (ex. 0, 1, 2)
reading_scores_by_grades = pd.DataFrame(reading_scores_by_grades).reset_index()
reading_scores_by_grades = reading_scores_by_grades.pivot(index = 'school_name', columns= 'grade',\
                                                          values = 'reading_score')
#REORDER COLUMNS to put in order
reading_scores_by_grades = reading_scores_by_grades[["9th","10th","11th", "12th"]]
reading_scores_by_grades
```

### SCORES BY SCHOOL SPENDING

```python
#BUDGET PER STUDENT (budget/# of students)
budget_per_student = school_data_complete["budget"]/school_data_complete["size"]
budget_per_student
```

```python
#CREATE NEW COLUMN IN SCHOOL_DATA_COMPLETE & ASSIGN IT TO THE SERIES
school_data_complete["Budget Per Student"] = budget_per_student
spending_bins = [0, 585, 615, 645, 675]
group_names = ["<$585", "$585-615", "$615-645", "$645-675"]
#Left side of equal = new column
#right side of equal = is what needs to be cut
school_data_complete["budget bin"]=pd.cut(school_data_complete["Budget Per Student"],\
                                          spending_bins, labels = group_names)
```

```python
#PUT VARIABLES IN A DICTIONARY & CONVERT TO DF
reading_score_school_spending = school_data_complete.groupby(["budget bin"]).mean()["reading_score"]
reading_score_school_spending 
```

```python
math_score_school_spending = school_data_complete.groupby(["budget bin"]).mean()["math_score"]
math_score_school_spending 
```

```python
#STUDENTS WHO DIDN'T PASS MATH
total_passing_math = school_data_complete.loc[school_data_complete["math_score"]>=70]
total_passing_math = total_passing_math.groupby(["budget bin"]).count()["Student ID"]
#STUDENTS WHO DIDN'T PASS READING
total_passing_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70]
total_passing_reading = total_passing_reading.groupby(["budget bin"]).count()["Student ID"]
#TOTAL NUMBER OF STUDENTS
total_students_2 = school_data_complete.groupby(["budget bin"]).count()["Student ID"]
```

```python
#DIVIDE STUDENTS WHO PASSED BY TOTAL NUMBER OF STUDENTS
#passing_math_score_school_spending = school_data_complete.groupby.count(["budget bin"]).mean()\
#["math_score"/"total_students"]
#passing_reading_score_school_spending = school_data_complete.groupby.count(["budget bin"]).mean()\
#["math_score"/"total_students"]
```

```python
#AVERAGE PASSING MATH
average_math_passing = total_passing_math/total_students_2*100
#AVERAGE PASSING READING
average_reading_passing = total_passing_reading/total_students_2*100
```

```python
overall_passing_score = school_data_complete.loc[school_data_complete["reading_score"]>=70]
overall_passing_score = overall_passing_score.groupby(["budget bin"]).count()["Student ID"]
overall_passing_score
overall_passing_rate = overall_passing_score/total_students_2*100
```

```python
school_spending = pd.DataFrame({
    "Average Math Score": total_passing_math,
    "Average Reading Score": total_passing_reading,
    '% Passing Math': average_math_passing,
    '% Passing Reading': average_reading_passing,
    "Overall Passing Rate": overall_passing_score})
    
#reorder columns
school_spending = school_spending[[
    "Average Math Score",
    "Average Reading Score",
    '% Passing Math',
    '% Passing Reading',
    "Overall Passing Rate"
]]
school_spending.index.name = "Spending Ranges (Per Student)"
```
